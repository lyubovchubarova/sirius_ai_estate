```python
from sklearn.cluster import AffinityPropagation
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Moscow_dataSet.csv', index_col=False)
df.info()
```

```python
AP = AffinityPropagation(damping=0.9, random_state=17).fit(df)
AP
```

```python
l = []
X = AP.labels_
for i in X:
    if i not in l:
        l.append(i)
print(min(l), max(l), '\n', l, '\n', sorted(l))
```

```python
ds = pd.read_csv('dataset.csv', index_col=False, names=['underground', 'length', 'latitude', 'floorscount', 'materialtype', 'buildyear', 'kitchenarea', 'roomscount', 'id', 'totalArea', 'offerType', 'floornumber', 'livingArea', 'price'], encoding='utf-8', delimiter=';')
```

```python
ds = ds.drop(['buildyear'],axis=1)
ds
```

```python
metrostation = []
mtypes = []
offertype = []
for i in dst['underground']: metrostation.append(i) if i not in metrostation else i
for i in dst['materialtype']: mtypes.append(i) if i not in mtypes else i
for i in dst['offerType']: offertype.append(i) if i not in offertype else i

print(f"{metrostation}\n{mtypes}\n{offertype}")
```
